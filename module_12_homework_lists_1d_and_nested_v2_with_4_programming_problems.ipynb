````python
# TODO: Implement F1 here

items = ["apples", "blueberries", "bananas"]

# adds an item to the list
def add_item(items, item):
    items.append(item)

# removes an item from the list
def remove_item(items, item):
    items.remove(item)

# prints a list of the items
def print_items(items):
    for i, item in enumerate(items):
        print(f"{i+1}. {item}")

# writes the items into shopping_lst.txt
def save_items(items, filename):
    with open(filename, "w") as f:
        for item in items:
            f.write(item + "\n")

# returns a list
def load_items(filename):
    list = []
    with open(filename, "r") as f:
        for line in f:
            list.append(line.strip())
    return list

if __name__ == "__main__":

    add_item(items, "pear")
    remove_item(items, "bananas")
    print_items(items)
    save_items(items, filename="shopping_lst.txt")
    load_items(filename="shopping_lst.txt")
````

### F2. Grade Statistics (1D list of numbers + file input)
You will read a file of grades (one number per line) into a 1D list, then compute statistics.

**Requirements:**
- Read `grades.txt` into a list `grades`
- Compute and print: count, min, max, average
- Create a new list `curved` that adds +5 to each grade (cap at 100)
- Write `curved` to `curved_grades.txt` (one per line)

**If you don't have a `grades.txt` file, create one in code first** (write 8–12 sample grades).


````python
# TODO: Implement F2 here
grades = []

# reads grades into a list
with open("grades.txt", "r") as f:
    for line in f:
        grades.append(line)

count = len(grades)
min_nums = min(grades)
max_nums = max(grades)
total = 0

# calculates the total of all the grades
for grade in grades:
    grade = int(grade)
    total += grade

print(f"Count: {count}")
print(f"Min: {min_nums}")
print(f"Max: {max_nums}")
print(f"Total: {total}")
print(f"Average: {total/count:.2f}")

curved = []

# curves the grades by 5 points if the value will not exceed 100
for grade in grades:
    grade = int(grade)
    if grade <= 95:
        grade += 5
        curved.append(grade)
    else:
        curved.append(grade)

# writes the curved grades into curved_grades.txt
with open("curved_grades.txt", "w") as f:
    for item in curved:
        item = str(item)
        f.write(item)
        f.write("\n")
````

### F3. Word Cleaner (file → 1D list → processing → file)
Read a text file (e.g., `paragraph.txt`) and build a 1D list of cleaned words.

**Requirements:**
- Read the file as text
- Convert to lowercase
- Remove punctuation characters like `. , ! ? ; :` (you may replace them with spaces)
- Split into words and store in a list `words`
- Build a list `long_words` containing only words with length ≥ 6
- Write `long_words` to `long_words.txt`, one per line

**If you don't have a `paragraph.txt`, create one in code first** (write 2–4 sentences).


````python
# TODO: Implement F3 here
long_words = []

with open("paragraph.txt", "r") as f:
    for line in f:
        line = line.lower().strip(".").strip(",").strip(";").strip(":").strip("!").strip("?")
        words = line.split(" ")

        # appends any word longer than 6 chars to a list
        for word in words:
            if len(word) >= 6:
                long_words.append(word)

# writes the words from teh long_words list to a file
with open("long_words.txt", "w") as f:
    for word in long_words:
        f.write(word + "\n")
````
